```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
# plt.rcParams['figure.figsize'] = 12,6  
```

```python
CO2_MaunaLoa = pd.read_csv('./data/greenhouse_gases/CO2_MaunaLoa_TimeSeries.csv',parse_dates=['Date'])
CO2_NOAA = pd.read_csv('./data/greenhouse_gases/CO2_NOAA_TimeSeries.csv',parse_dates=['Date'])
CO2_WDCGG = pd.read_csv('./data/greenhouse_gases/CO2_WDCGG_TimeSeries.csv',parse_dates=['Date'])
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(CO2_MaunaLoa['Date'],
         CO2_MaunaLoa['Mauna Loa CO2 (ppm)'],
         linewidth=2,label='Mauna Loa',
         color='black')

plt.suptitle('Mauna Loa CO2 Conentrations')
## plot a legend on the ax
plt.legend()
```

```python
CO2_NOAA
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(CO2_NOAA['Date'],
         CO2_NOAA['NOAA CO2 (ppm)'],
         linewidth=2,label='NOAA',
         color='black')

plt.suptitle('NOAA CO2 Conentrations')
## plot a legend on the ax
plt.legend()
```

```python
CO2_WDCGG
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(CO2_WDCGG['Date'],
         CO2_WDCGG['WDCGG CO2 (ppm)'],
         linewidth=2,label='WDCGG',
         color='black')

plt.suptitle('WDCGG CO2 Conentrations')
## plot a legend on the ax
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.plot(CO2_MaunaLoa['Date'],
         CO2_MaunaLoa['Mauna Loa CO2 (ppm)'],
         linewidth=2,label='Mauna Loa',
         color='black')

ax.plot(CO2_NOAA['Date'],
         CO2_NOAA['NOAA CO2 (ppm)'],
         linewidth=2,label='NOAA',
         color='red')

ax.plot(CO2_WDCGG['Date'],
         CO2_WDCGG['WDCGG CO2 (ppm)'],
         linewidth=2,label='WDCGG',
         color='blue')

plt.xlabel('Year')
plt.ylabel('CO2 Concentration (ppm)')
                       
plt.legend()
plt.suptitle('CO2 Conentrations from Mauna, NOAA, and WDCGG')
```

# Figure 1: CO2 concentrations from all 3 monitoring stations (Mauna Loa, NOAA, WDCGG)
As the plot shows, Mauna Loa began its monitoring in the 1950's, and the other two stations began monitoring in the 1980's. During the period of time where all three stations are monitoring, you can clearly see that they are recording nearly identical values.
